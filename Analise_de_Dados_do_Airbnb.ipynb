A partir dessas observações, identificamos áreas que podem enriquecer nossa análise, visando a eliminação de outliers. Com o intuito de aprofundar nossa compreensão diante dos dados disponíveis, planejamos empregar o método do escore IQR. Essa abordagem nos auxiliará na identificação de valores discrepantes, permitindo estabelecer limites superiores e inferiores. Nossa intenção é aplicar essa faixa de valores, tanto para os preços quanto para o mínimo de noites, com o propósito de aprimorar a integridade da nossa análise.

O IQR score é uma maneira de ajudar a entender a dispersão dos dados, especialmente quando temos informações incomuns ou diferentes em um conjunto de números. IQR significa "Intervalo Interquartil", que é a diferença entre o terceiro quartil (75%) e o primeiro quartil (25%) de um conjunto de dados. Em outras palavras, ele mostra a faixa onde a maior parte dos dados está concentrada.

<center><img src="https://editor.analyticsvidhya.com/uploads/15066IQR-Boxplot.png"><center/>

Para calcular o IQR score, você subtrai o primeiro quartil dos dados do terceiro quartil. Depois, você multiplica esse valor por um número chamado "fator de outlier". Esse fator ajuda a definir um limite superior e inferior além dos quais os valores são considerados incomuns ou atípicos.

Então, o IQR score é uma maneira de identificar valores estranhos ou diferentes em seus dados, ajudando a entender se existem valores que se destacam da maioria. Isso pode ser útil para saber se há informações que não seguem o padrão geral dos outros dados.

```
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
iqr_price = q3_price - q1_price
min_iqr = q1_price - (1.5*iqr_price)
max_iqr = q3_price + (1.5*iqr_price)

print(f'O nosso  Q1 tem o valor de {q1_price}')
print(f'O nosso  Q3 tem o valor de {q3_price}')
print(f'O nosso  IQR score tem o score de {iqr_price}')
print(f'O nosso outlier inferior é {min_iqr}')
print(f'O nosso outlier superior é {max_iqr}')
```

Agora com esses nossos resultados do IQR podemos ver se ainda temos muitos outliers ou poucos em relação ao preço

```
fig, ax = plt.subplots(figsize=(15,3))
df.price.plot(kind="box", vert=False)
ax.set_title('Datframe com outliers de preço')
plt.show()
print(f'Quantidade de Entradas : {df.shape[0]}')
print(f'Os valores acima de 120 euros tem {len(df[df.price>120])} entradas')
print('repesentam {:.2f}% do Dataframe original'.format((len(df[df.price>120])/df.shape[0])*100))
```

Vejamos que temos 743 outliers que representa `24.92%` do nosso Dataframe original e vemos que nosso IQR faz sentido, vamos ver agora se o minimo de noite também tem o mesmo resultado positivo que o preço teve

```
q1_mg = df.minimum_nights.quantile(.25)
q3_mg = df.minimum_nights.quantile(.75)
iqr_minimun_nights = q3_mg - q1_mg
min_minimun_nights = q1_mg -(1.5*iqr_minimun_nights)
max_minimun_nights = q3_mg +(1.5*iqr_minimun_nights)

print(f'O nosso  Q1 tem o valor de {q1_mg}')
print(f'O nosso  Q3 tem o valor de {q3_mg}')
print(f'O nosso  IQR score tem o score de {iqr_minimun_nights}')
print(f'O nosso outlier inferior é {min_minimun_nights}')
print(f'O nosso outlier superior é {max_minimun_nights}')
```
